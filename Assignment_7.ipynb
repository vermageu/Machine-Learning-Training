```python
#1. Import the dataset and write a description about the dataset. 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("C:/Users/Intel/Desktop/titanic.csv")
print("On April 15, 1912, the largest passenger liner ever made collided with an iceberg during her maiden voyage. When the Titanic sank it killed 1502 out of 2224 passengers and crew.")
print("\nThe titanic.csv dataset contains data for 887 of the real Titanic passengers.")
print("\nEach row represents one person. The columns describe different attributes about the person including whether they survived (S), their age (A), their passenger-class (C), their sex (G) and the fare they paid (X). ")
```

```python
#2. Find the count of the missing values and describe how you have handled the missing data. 
print("\nTotal Null Values: ")
print(sum(df.apply(lambda x: sum(x.isnull().values), axis = 1)))
print("The null value is handled by isnull() function. Consequently, we have no null value in this dataset.")
```

```python
#3. Univariate Analysis - Plot the univariate graphs and find the outliers.
data = df["Fare"]
sns.boxplot(data)
print("The outliers are clearly visible with the help of Boxplot\n\n")
```

```python
data = df["Age"]
sns.boxplot(data)
print("Most of the people onboard were between 20-38 years old. People above 65 can be seen as outliers in the boxplot\n\n")
```

```python
#4. Bivariate Analysis - Plot the bivariate graphs and find the relationship between the variables. 
print(pd.crosstab(df['Sex'],df['Survived'],margins='True'))
sns.countplot(x='Sex',hue='Survived',data=df)
print("\t\t::FINDINGS::")
print("The number of women survived was more than that of men.\nMen were more in numbers onboard.")
```

```python
#5. Multivariate Analysis - Plot the multivariate graphs and find the relationship between the variables. 
sns.pairplot(df)
```
